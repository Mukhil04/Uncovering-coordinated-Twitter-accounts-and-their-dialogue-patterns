```python
cluster_emb = []
cluster = []
with open('u_cluster.npy') as f:
    lines = f.readlines()
    for val in lines:
        cluster.append(int(val[0]))
with open('u_emb.npy') as f:
    lines = f.readlines()
    for values in lines:
        score = values[:-1].split(" ")
        temp = []
        for val in score:
            temp.append(float(val))
        cluster_emb.append(temp)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
a = pca.fit(cluster_emb)
print(a.explained_variance_ratio_)
result = pca.fit_transform(cluster_emb)
```

```python
import matplotlib.pyplot as plt
red_points_x = []
red_points_y = []
red_points_z = []
blue_points_x = []
blue_points_y = []
blue_points_z = []
for i in range(len(cluster)):
    if cluster[i] == 0:
        blue_points_x.append(result[i][0])
        blue_points_y.append(result[i][1])
        blue_points_z.append(result[i][2])
    elif cluster[i] == 1:
        red_points_x.append(result[i][0])
        red_points_y.append(result[i][1])
        red_points_z.append(result[i][2])
        
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.scatter3D(blue_points_x, blue_points_y, blue_points_z, c='b')
ax.scatter3D(red_points_x, red_points_y, red_points_z, c='r')
# plt.scatter(blue_points_x, blue_points_y, c='b')
# plt.scatter(red_points_x, red_points_y, c='r')      
plt.show()
# Less red and more blue
```
